```python
set = {1,2,3,5}
print(set)
print(type(set))
# Here the set is printed and the type of set is also printed 
```

```python
s1 = {1,2,4,5,4,1,6,7,8,9,2,4,1}
print(s1)
print(type(s1))
# Here the set is converted in order with no repetition 
```

```python
s3 = ({5,6,7})
print(s3)
# The se s3 gets printed 
```

```python
# Add 
s = set()

s.add(5)
s.add(13)
s.add(12)
s.add(5)

print(s)
```

```python
s1 = {3,9,7,4}
s1.discard(7)
print(s1)
# Here as we can see that the number 7 gets discarded from the set 
```

```python
s1 = {3,9,7,4}
s1.remove(7)
print(s1)
# Here as we can see that the remove function gives the same result 
```

```python
# froxenset 
# A frozenset is an immutable set 

fs =frozenset([1,2,3])
print(fs)
print(type(fs))
```

```python
fs2 = frozenset({1,2,3})
print(fs2)
print(type(fs2))
# here the frozenset is shown here 
```

```python
dict1 = {fs2 : "Vishwa"}
print(dict1)
# Frozenset used with dicionary 
```
