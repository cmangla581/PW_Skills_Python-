```python
# String Concatenation 
first_name = 'Chaitanya'
last_name =  'Mangla'
complete_name = first_name + last_name
complete_name
```

```python
# Also multiple strings can be concatenated 
# here we want space to be added between the 2 names of the strings 
first_name = 'Chaitanya'
last_name = 'Mangla'
complete_name = first_name + " " + last_name
complete_name
```

```python
print(len(complete_name))
# here as we can see that the length of the complete name gets printed on using len function 
```

```python
# Converting the string to uppercase 
str = "Vishwa Mohan"
print(str.upper())
```

```python
# Converting the string to lower case 
str = "VISHWA MOHAN"
print(str.lower())
```

```python
# Capitalize Function 
str = "vIshwa Mohan"
print(str.capitalize())
# So here it prints the first letter of the string as capital and rest are all converted to small letters 
```

```python
# Strip Function 
str = " vIshwa"
print(str.strip())
# Here as we can see that the gap will get vanished away 
```

```python
# Replace Function 
str = "Hello dear, I am very good, I am very nice, I am wow"
print(str.replace("I am", "We are"))
# Here as we can see that i am gets replced by We are 
```

```python
# Replace Function 
str = "Hello dear, I am very good, I am very nice, I am wow"
print(str.replace("I am", "We are", 2))
# here since we want to replce only 2 times, so we use this function 
```

```python
# Escape Character 
print("Hello\nVishwa")
# hello and Vishwa printed on 2 different lines 
```
