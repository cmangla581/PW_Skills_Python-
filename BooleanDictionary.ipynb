```python
students_details = {'name':'Vishwa', 'age':'21', 'city':'Bangalore'}
students_details['country'] = 'India'
print(students_details)
```

```python
# Updating the details of students 
students_details = {'name':'Vishwa', 'age':'21', 'city':'Bangalore'}
marks_details ={'English' : 100, 'Maths': 78, 'Science':92}

students_details.update(marks_details)
print(students_details)
# Here it is showing that the marks is added into the dictionary 
```

```python
# Removing od Students Details from the whole dictionary 
# Pop Function
print(students_details.pop('city'))
print(students_details)
```

```python
# popitem Function 
print(students_details.popitem())
print(students_details)
# Pop item function is used to pop the function which is at the last of the dictionary 
```

```python
# Delete Function 
del students_details['age']
print(students_details)
# Here as we can see that the age details are deleted in the fnction 
```

```python
# Clear Function 
students_details.clear()
print(students_details)
# it will print an empty function call value 
```
