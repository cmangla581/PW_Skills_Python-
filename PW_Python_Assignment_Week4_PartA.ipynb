Q8: Define a function that accepts a variable number of argumentsusing "args" and demonstrate its usage with example ?

Ans. The args syntax allow a function to accept any number of positional arguments, which are then accessed as a tuple inside the
      function. 

```python
def sum_numbers(*args):
    total = 0
    for num in args:
        total += num
    return total

print(sum_numbers(1,2,3,4,5))    

# Sum gets printed as the output 
```

Q14. What is Call by Value and Call  y Reference ? Explain with exaamples ? 

Ans.  
    Call by Value :  InCall by value, a copy of variable's value is passsed to the function. 
                    Any modifications, made to the parameter inside the function do not affect the original variable. 

    Call by Reference : In call by Reference, a referenc e to the variable is passed to the function.
                         Any modification, made to the parameters inside the function affect the global variable. 
                         

```python
# Call by value
def increment (num):
    num += 1 
    print("in inside function Value of x", num) 

x = 10 
increment(x)
print("In outside function value of x", x)
```

```python
# Call by Rference 

def modify_list(ist):
    ist.append(4)
    print("inside Function:", ist) 

my_list = [1,2,3]
modify_list(my_list)
print("Outside function", my_list)
```
