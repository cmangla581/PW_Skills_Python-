Q8. PW store plans to give PW Vishwas Diwas Offer 2024 discounts to its Students. The store management has decided to give discount on the following criteria:
shopping_amount >= 2500 and shopping_amount < 3000: discount_rate =5%
shopping_amount >= 3000 and shopping_amount < 4500:  discount_rate =8%
shopping_amount >= 4500:  discount_rate =10%

An additional discount of 5% is given to customers who are the members of the store. Create a program using a user defined function that accepts the shopping amount as a parameter and calculates discount and net amount payable on the basis of the following conditions:
Net Payable Amount = Total Shopping Amount – Discount.


```python
def calculate_discount(shopping_amount,is_member):
    discount_rate = 0

    # Check if the customeer is a member 
    if is_member:
        discount_rate += 0.05

    if shopping_amount >= 2500 and shopping_amount < 3000 :
        discount_rate += 0.05
    elif shopping_amount >= 3000 and shopping_amount < 4500 : 
        discount_rate += 0.08 
    elif shopping_amount >= 4500:
        discount_rate += 0.10

    # Calculate discount amount 
    discount_amount = shopping_amount * discount_rate

    # Calculate net payable amount 
    net_amount_payable = shopping_amount - discount_amount

    return discount_amount, net_amount_payable

shopping_amount = float(input("Enter the shopping amount: "))
is_member = input("Are you a member of the store (yes/no)").lower() == "yes"

# Call the function to calculate the discount and net payable amount  
discount_amount , net_amount_payable = calculate_discount(shopping_amount, is_member)

# Display the results 
print("Discount Amount: INR{:.2f}".format(discount_amount))
print("Net Payable Amount: INR{:.2f}".format(net_amount_payable))


# Amount = 25000 
```

```python
def calculate_determinant(a,b,c):
    determinant= lambda a,b,c : b**2 - 4*a*c
    return determinant(a,b,c)

# Input Coefficients from the user 
a= float(input("Enter the coefficient of x^2 (a)"))
b= float(input("Enter the coefficient of x (b)")) 
c = float(input("Enter the constant term (c)")) 

determinant =  calculate_determinant(a,b,c)

if determinant > 0 :
    print('the quadratic equatio has two distinct reaal roots')

elif determinant == 0:
    print("The quadratic equation has one real root")

else:
    print("the quadratic equationhas two complex roots")

print("Determinant", determinant) 

# a = 3, b= 4, c = 5 
```

Q12. Write a program to calculate the total cost of the canvas used for making a tent and the net amount payable, including tax, using user-defined functions.

Hint:
Calculate the Area of the Cylindrical Part of the Tent:
area_cyl=2×π×radius×height.
Calculate the Area of the Conical Part of the Tent:
 Use the formula area_con =π×radius×slant height.
Calculate the Total Area of Canvas Used for Making the Tent: Add the areas of the cylindrical and conical parts.
Calculate the Total Cost of the Canvas Before Tax: Multiply the total canvas area by the unit price(1 Rupee per unit).
Compute the Net Amount Payable for the Tent, Including Tax: Calculate tax as 18% of the cost. Add the tax to the cost to get the net price.
Print the Results: Print the total cost of the canvas before tax and the net amount payable (including tax).


```python
# program to calculate the cost of the tent 
# Function Definition 

def cyl(h,r):
    area_cyl =  2 * 3.14 *r*h
    return area_cyl 

def con(l,r):
    area_con = 3.14 * r * l 
    return area_con

def post_tax_price(cost):

    tax = 0.18 * cost 
    net_price = cost + tax 
    return  net_price 

print("Enter values of the cylindrical part of the tent in meters: ")
h = float(input("Height (H): "))
r = float(input('Radius (R)'))
csa_cyl = cyl(h,r)   # Function Call 

print('Enter the slant height of the conical area in meters')
l = float(input("Slant Height (l):  "))
csa_con = con(l,r) 

canvas_area = csa_cyl + csa_con 
print("Area of canvas: ", canvas_area, " m^2") 

# Calculae the cost of canvas 

unit_price = float(input("Enter cost of 1 m^2 cnvas in rupees:  "))
total_cost = unit_price * canvas_area

print("total cost of canvas before tax = ", total_cost)
print("Net amount payable = ", post_tax_price(total_cost))


# Height = 22, Radius = 21, Slant Height = 24, Cost = 12 

```
