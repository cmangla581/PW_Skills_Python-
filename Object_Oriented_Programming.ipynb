                                  Methods and its Types

```python
# Instance Method 
class Person : 
    def __init__(self, name, age):
        self.name = name 
        self.age = age 

    def findAge(self):
        return self.age
    
per = Person("Vishwa", 99)
print(per.findAge()) 

# Here as we can see tht on applying the code, we can find the age from the function 
```

```python
# class Method 
# Any method which is used at class level is called class method 

class Person :

    country = "India"

    classmethod
    def greet(cls):
        print("Hello from the ", cls.country)

    
    def __init__(self, name, age):
        self.name = name 
        self.age = age 

    def findAge(self):
        return self.age
    
per = Person("Vishwa", 99)
print(per.findAge()) 

per.greet() 
```

```python
class Person :

    country = "India"

    @classmethod
    def greet(cls):
        print("Hello from the ", cls.country)

    @staticmethod
    def hello():
        print("Hello")

    
    def __init__(self, name, age):
        self.name = name 
        self.age = age 

    def findAge(self):
        return self.age
    
per = Person("Vishwa", 99)
print(per.findAge()) 

per.greet() 
```

                                          Access Modifier

There are 2 types of Access Modifiers : Public and Private 

```python
# Example of the access modifier by code 

class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age 

        def findAge(self):
            return self.age 
        

per = Person("Vishwa", 99)
print(per.name)

# By default, this is Public 
```

```python
class Person : 
    def __init__(self,name,age,salary):
        self.name = name
        self.age = age 
        self.__salary = salary 

    def findAge(self):
        return self.age
    
    def getSalary(self):
        print(self.__salary)


per = Person("Vishwa", 99, 5000)
print(per.name)
print(per.__salary)

# Here we can see the name bu the salary is not visible as because its made private 
```

```python
class Person : 
    def __init__(self,name,age,salary):
        self.name = name
        self.age = age 
        self.__salary = salary 

    def findAge(self):
        return self.age
    
    def getSalary(self):
        print(self.__salary)


per = Person("Vishwa", 99, 5000)
print(per.name)
per.getSalary()

# Here the name along with salary is written visible on them 
```

                                        Encapsulation

Encapsulation is basically an object oriented programming fundamental principle, that involves bundling
of data attributs and methods that operate on data. 

The attributes must be private. 

```python
# Encapsulation 
class Person: 
    def __init__(self, name, car):
        self.__name = name
        self.__car = car

    def getName(self):
        return self.__name
    
    def setName(self,name):
        self.__name = name

    def getCar(self):
        return self.__car 
    
    def setcar(self, car):
        self.__car = car

per = Person("Vishwa", 99)

print(per.getName())
per.setName("Vishwa Mohan Singh")
print(per.getName())

# This is known as the setter and getter method 
```

                                       Inheritance and its Types

Inheritance in OOP is that one classs acquires the properties and behaviours of theanother class. 

Inheritance is a core mechanism where one class acquires  proerties and behaviours from another class. 

```python
class Animal:
    def __init__(self,name):
        self.name= name

    def eat(self):
        print(self.name + "Animal is eating")

class Dog(Animal):
    def __init__(self,name,type):

        Animal.__init__(self,name)
        self.type = type

    def getTheNameofDog(self):
        print(self.name)

dog = Dog("Moti", "Doberman")
dog.eat()

dog.getTheNameofDog()

# Hnce we get the name of the dog 
```

```python
# Example of the code 

class Parent : 
    property= 90 

class Child(Parent):
    property = 99 

    def display(self):
        print("Child Property", self.property)
        print("Parent property", super().property)

obj = Child()
obj.display()
```

3 types of inheritance : 
1. Single Level Inheritance
2. Multi Level Inheritance 
3. Hierarchial Inheritance 

```python
# Example of Hierarchial Inheritance  

class Animal : 
    def __init__(self,name):
        self.name = name 

    def eat(self):
        print("Animal is eating")

class Dog(Animal):
    def __init(self,name,type):
        super().__init__(name)
        self.type = type

class Cat(Animal):
    def __init__(self,name, type):
        super().__init__(name)
        self.type = type 
```

```python
# Example of Multi inheritance 
class A : 
    def meth1(self):
        print("Hello from A") 

class B : 
    def meth1(self):
        print("Hello from B")

class C (A,B):
    def meth(self):
        print("Hello from the child")


c = C()
c.meth()
```

```python
# Hybrid inheritance 

class Pet(Dog, Cat):
    pass

#Combination of Dog and Cat 
```

                                       Abstraction

Abstraction is a core principle in OOP that simplifies complex systems by hiding un necesssary implemenations details 
 and showing only essential features. 

```python
# Abstraction Method 

class Animal : 
    def eat(self):
        pass 

animal = Animal()
animal.eat()
```

```python
from abc import ABC,abstractmethod 

class Animal(ABC):
    @abstractmethod
    def eat(self):
        pass 

a = Animal()
```

```python
class Dog(Animal):

    def sleep(self):
        print("Dog is sleeping")

d = Dog()

# Method overriding hapens when both parent and child has same method 
```

```python
# Example of Polymorphism 

class Animal : 
    def eat(self):
        print("Animal is eating")

class Dog(Animal) : 
    def eat (self):
        print('Dog is eating')

class Cat(Animal):
    def eat (self):
        print("Cat is eating")

c = Cat()
d= Dog()
a= Animal()

c.eat()
d.eat()
a.eat()
```

```python
# Another example of Polymorphism 

class Birdfly : 
    def flyBird(self, bird):
        bird.fly()

class Parrot :
    def fly(self):
        print("Parrot is Flying")

class Crow : 
    def fly(self):
        print("Crow is flying") 

p = Parrot()
c = Crow() 

bf = Birdfly()

bf.flyBird(p)
bf.flyBird(c) 

```
