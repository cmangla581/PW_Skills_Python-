Tasks:
1. Define base class(es) for bank accounts with common attributes and methods.
2. Implement subclasses for specific account types (e.g., SavingsAccount, CheckingAccount).
3. Provide methods for deposit, withdraw, and balance inquiry in each subclass.
4. Test the banking system by creating instances of different account types and performing transactions.


```python
# The code for the following bank system is as : 

class BankAccount: 
    def __init__(self, account_number, balance=0): 
        self.account_number = account_number 
        self.balance = balance 


    def deposit(self,  amount): 
        if amount > 0 : 
            self.balance += amount 
            print(f"Deposited ${amount}. Current balance: ${self.balance}") 

        else : 
            print("Invalid Deposit Amount") 

    def withdraw (self,amount): 
        if 0 < amount <= self.balance :
            self.balance -= amount 
            print(f"Withdraw ${amount}. Current Balance: ${self.balance}") 

        else : 
            print("Insufficient funds or invalid withdrawl amount") 

    def inquire_balance(self): 
        print(f"Account Balance: ${self.balance}") 


class SavingsAccount(BankAccount): 
    def __init__(self, account_number,balance = 0, interest_rate = 0.01): 
        super().__init__(account_number, balance) 
        self.interest_rate = interest_rate 

    def add_interest(self): 
        interest = self.balance * self.interest_rate
        self.balance += interest 
        print(f"Interest Added. Current Balance : ${self.balance}") 

class CheckingAccount(BankAccount): 
    def __init__(self, account_number, balance = 0, overdraft_fee = 25): 
        super().__init__(account_number, balance) 
        self.overdraft_fee = overdraft_fee

    def withdraw(self, amount): 
        if amount > self.balance : 
            print("Overdraft. Appllying Overdraft Fee ")
            self.balance -= self.overdraft_fee 
        super().withdraw(amount) 


# Testing the bank system  

def main(): 
    savings_acc = SavingsAccount ("PUNB001", 1000, 0.01)
    checking_acc = CheckingAccount ("UCO009", 500) 

    #Test deposit, withdrw and balance_enquiry 

    savings_acc.deposit(1500) 
    savings_acc.add_interest()
    savings_acc.withdraw(100) 
    savings_acc.inquire_balance() 

    checking_acc.deposit(1200)
    checking_acc.withdraw(1700)
    checking_acc.inquire_balance() 

if __name__ == "__main__" : 
    main()


```

```python
# Ths code for the above question is :  

class Room: 
    _next_room_number = 201

    def __init__(self, room_type, rate): 
        self._room_number = Room._next_room_number 
        self._room_type = room_type 
        self._rate = rate 
        self._is_available = True 
        Room._next_room_number += 1 

    def get_room_number(self): 
        return self._room_number 
    
    def get_room_Type(self): 
        return self._room_type 
    
    def get_rate (self): 
        return self._rate 
    
    def is_available(self): 
        return self._is_available 
    
    def book_room(self): 
        if self._is_available: 
            self._is_available = False 
            print(f"Room {self._room_number} ({self._room_type}) has been booked") 

        else : 
            print(f"Room {self._room_number} ({self._room_type}) is not available")

    def check_in(self): 
        if not self._is_available: 
            print(f"Guest checked in to Room {self._room_number} ({self._room_type})")

        else : 
            print(f"Room {self._room_number} ({self._room_type}) is not booked ")

    def check_out(self): 
        if not self._is_available: 
            self._is_available = True 
            print(f"Guest checked out of the room {self._room_number} ({self._room_type})") 

        else : 
            print(f"Room {self._room_number} ({self._room_type}) is already available") 


class SuiteRoom(Room): 
    def __init__(self, rate, has_jacuzzi): 
        super().__init__("Suite", rate)
        self._has_jacuzzi = has_jacuzzi 

    def get_has_jacuzzi (self): 
        return self._has_jacuzzi 
    
    def set_has_jacuzzi(self, has_jacuzzi): 
        self._has_jacuzzi = has_jacuzzi 

class StandardRoom(Room): 
    def __init__(self, rate,has_balcony): 
        super().__init__("Standard", rate)
        self._has_balcony = has_balcony 

    def get_has_balcony(self):
        return self._has_balcony 
    
    def set_has_balcony(self, has_balcony): 
        self._has_balcony = has_balcony 


# Example Usage  
def main(): 
    suite_room = SuiteRoom(200.0,  True)
    standard_room = StandardRoom(150.0, False) 

    #  Book and check the in/ out rooms 
    suite_room.book_room()
    suite_room.check_in()
    suite_room.check_out() 

    standard_room.book_room()
    standard_room.check_in()
    standard_room.check_out()

if __name__ == "__main__": 
    main()
```

```python
# The code for the above question is as : 

class ShoppingList : 
    def __init__(self): 
        self.cart = []
        self.total_price = 0.0 

    def add_item(self, item_name, price, quantity= 1): 
        item = { 
            "name" : item_name,
            "price" :  price, 
            "quantity" : quantity 

        }
        self.cart.append(item)
        self.total_price +=  price * quantity 

    def remove_item(self, item_name): 
        for item in self.cart:
            if item["name"] == item_name:
                self.total_price =item["price"] * item["quantity"]
                self.cart.remove(item)
                print(f"{item['name']} has been removed from the cart")
                return 
        
        print(f"{item_name} is not in the cart")


    def calculate_total(self): 
        return self.total_price
    
    def display_cart(self): 
        if not self.cart:
            print("Your cart iss Empty")

        else : 
            print("Items in your crt: ") 
            for item in self.cart: 
                print(f"{item ['name']} - ${item['price']}  x {item['quantity']}")
            print(f"Total: ${self.total_price:.2f}")

# Example usage 

cart = ShoppingList()

cart.add_item("Apple", 0.5, 3)
cart.add_item("banana", 0.5,2)
cart.add_item("Milk", 3.99, 1)

cart.display_cart()

cart.remove_item("Banana")

cart.display_cart()

print("Total Price: ",  cart.calculate_total())
```

Q 12. 
Write a Python program to create a calculator class. Include methods for basic arithmetic operations.



```python
class Calculator : 
    def add(self, a,b): 
        return a +b 
    
    def subtract(self, a,b): 
        return a-b
    
    def multiply(self,a,b): 
        return a*b
    
    def divide (self, a,b): 
        if b == 0 : 
            return "Error"
        return a/b
    

# Example usage 

calculator = Calculator() 

print(calculator.add(5,3))
print(calculator.subtract(5,3))
print(calculator.multiply(5,3))
print(calculator.divide(5,3))
print(calculator.divide(5,0))
```
