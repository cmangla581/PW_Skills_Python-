                         Variables in Functions (Global and Local Scope)

```python
x = 101  # Its a Global Scope

def func():
    x = 102   # Its a local Scope 
    print(x) 

func()
print(x)

```

Global Variable : 
A variable defined outside the function is called as the Global Variable.  

Local Variable : 
A Variable defined inside the function is called the Local Variable and it doesn't affect the overall function. 
Its terminated as soon as the functions execution gets completed. 

                                        Different Types of Arguments

1. Default Argument

```python


def greet (name, message):
    print(name,  message)

greet ("Vishwa", 'Hello')
# Here as we can see that the name and message getting printed normally 

```

```python
# Now, we define the function by using the default argument method  
def greet (name, message = "Good Morning"):
    print(name, message) 

greet("Vishwa") 

# Here in the default argument, Vishwa Goood morninng gets printed by defualt due to defaullt argument 
```

2. Keyword Arguments 

```python
def greet(name, age, message):
    print(message, name, "your age is",age)

greet(age = 99, name = 'Chaitanya', message = "Good morning")  

# The benefit of the Keyword arguments is that by using the keywords, we can also swap the order of arguments 
```

3. Positional Arguments 

```python
def add_numbers(x,y):
    print("X:", x)
    print("Y:", y)
    print(x + y) 

add_numbers(6,5)

# Here as we can see that in positional arguments the 6 and 5 get printed and added in a specific order 
```

Suppose, we have to define a function of adding multipe numbers 
then we apply the technique as

```python
def sum_numbers(*args):
    print(type(args))
    print(args)

    sum = 0 
    for num in args :
        sum += num
    return sum 

print(sum_numbers(1,2,3,4,5))

# Here as we can see that the class tupl gets printed and the sum 15 is also printed  
```

```python
# Now, suppose we have to define a function 

def fn (a,b,*args):
    print(a)
    print(b)
    print(args)
    print(*args)

fn(5,6,7,8,9)

# Here as we can see that 5 and 6 gets printed for a and b 
# Whereas, the numbers 7,8,9 get formed as the tuples 
# Here the symbol * is used for removing the brackets of the tuples 
```

                                                 Kwargs in Python

```python
# Example of code for Kwargs 

def display_info(**kwargs):
    print(kwargs)
    print(type(kwargs))

    for key , value in kwargs.items():
        print(key , " -> ", value)

display_info(name = "Vishwa", age = 99, city = "Bangalore")
```

```python
# One more example of code in kwargs  

def func (a,b,*args , **kwargs):
    print(a)
    print(b)
    print(args)
    print(kwargs)

func(5,6,7,8,9, name = "Vishwa" , age = 99) 

# Here, as we can see that the positional argumentation takes place and : 5,6 in a and b , 7,8,9 in args  ,  Vishwa and 99 in kwargs
```

```python
def func1 (**kwargs , a, b): 
    print(kwargs)
    print(a)
    print(b)

func1(name = 'Vishwa', age = 99, 7,9)

# Here it shows error as the kwarg shuld always be the last parameter 
```

                                        Return Types and Function Nesting 

```python
# Example of Code  

def add_numbers(a : int, b : int) -> int:
    return a + b

print(add_numbers(5,6))
print(add_numbers(5.5,6.5)) 

# Here this is known as implicit conversion. Although we are passing the integer paramenters, but also working on the decimal arguments
```

```python
# Example of function nesting  
def outer ():
    print("Hello from outside")

    def inner():
        print("Hello from inside")

    return inner  

```

```python
fn = outer()
```

```python
fn()
```

```python
outer()()
```

                              Pass by Value & Pass by Reference

Pass by value is a parameter passing method where a function receives a copy of the actual argumen value
and not the original value itself. 

```python
# Example of pass by value using a code 

num = 5 
def modify_num(num):
    num += 1
    print(num)

modify_num(num)

print("Original Num", num)

# So here we are writing the copy of the number and then 
```

Immutable Data Types : numbers, Strings and Tuples -> pass By Value 

Pass by Reference -> Mutable Data Types, List, Dictionary  

```python
my_list = [1,2,4]

def modify_list(l1):
    l1.append(5)
    print(l1)

print("Before Calling fun", my_list) 

modify_list(my_list) 

print("After Calling fun", my_list)
```

Lambda Function : A lambda Function is a small annonymous function defined with the lambda Keyword that can take
any number of arguments but is restriicted to a single expression. 

```python
# Example of Lambda Function by a code 

func = lambda x : x + 10 
print(func(5))
# 15 gets printed as the result 

add = lambda a,b : a+b
print(add(5,6))
# 11 gets printed as a result 
```

```python
def myFunc():
    return lambda msg : print(msg)

myFunc()("Hello World") 
# So here this is how the function takes place  
```

```python
numbers = [23,45,67,89,7,2,1,87]

greater_than_9 = list(filter(lambda x : x > 9, numbers))

print("Numbers greater then 9 are:", greater_than_9)
```

```python
numbers = [23,45,678,54,12,9]

multiples_of_5 = list(filter(lambda x : x % 5 ==0, numbers))

print("Multiples of 5 are:", multiples_of_5)
```
