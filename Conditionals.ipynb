```python
# Now we use elif condition along with it 
x = int(input("Ener the number"))
if (x >15) : print("The number is greater than 15")
elif (x == 15) : print("The number is equal to 15")
else : print("The number is leass than 15")

# Here when we enter the number 15, it enters that the number is equal to 15 
```

```python
# Grading System with else, if and elif conditionals 
marks = int(input("Enter the marks"))

if (marks <= 100 and marks >= 91) : print("A")
elif (marks <= 90 and marks >= 81) : print("B") 
elif (marks <= 80 and marks >= 71 ) : print("c")
elif (marks <= 70 and marks >= 61) : print('D')
 
else : print("Failed, Sit at home")

# Input is 89 
```

```python
# Nested Conditions 
x = int(input("Enter a number "))

if (x > 5):
     print("x is greater then 5")
    
     if(x%2 == 0): print("x is even")
     else : print("x is odd")

else: print("X is less than 5, not calculating odd or even ")
# Input is 234
```

```python
# Match and case Statements 
# Display days of the week by match and case 

num = int(input("Enter a number "))
match num :
    case 1:
        print("Today is Monday")
    case 2:
      print('Today is Tuesday ')
    case 3:
      print("Today is Wednesday")
    case 4:
      print("Today is Thursday")
    case 5:
      print("Today is Friday")
    case 6:
      print("Today is Saturday")
    case 7:
      print("Today is Sunday") 
    case _:
      print("Invalid Day number")

# Input 6 
```
