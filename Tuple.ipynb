```python
# defining a Tuple 
tup1 = (3,6,1,9,10)
print(tup1)
# Here the tuple is getting printed 

print(type(tup1)) 
# hence the type tuple is getting printed 
```

```python
# We also have the tuple constructor method 
t1 = tuple([1,2,4])
print(t1)
# here the tuple gets printed using the tuple constructor method 
```

```python
t2 = tuple("vishwa")
t3 = tuple(tup1)
print(t2,t3)
# here as we can see that t2 and t3 get printed as seperate tuples 
```

```python
t =(35,)
print(t)
print(type(t))
# Here a comma presents that this is a tuplw 
```

```python
# Count method in tuple 
fruits = ('apple', 'mango', 'banana', 'apple', 'strawberry')
print(fruits.count('apple'))

print(fruits.count('jackfruit'))
# Sice no jackfruit is there so it will print 0 
```

```python
# Index Printing method 
print(fruits.index('apple'))
print(fruits.index('mango'))

# The index of both apple and miango will get printed 
```

```python
# printing the length of the fruits 
print(len(fruits))
```
