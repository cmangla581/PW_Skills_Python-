Loops is something which we use when we have to repeat something. 
Like for example, we have to print "Hello World" around 1000 times, then its obvious that we cannot 
write it continuosly in 1000 differetn print statements 

In such cases, we use loops  

```python
fruits = ["apple", "banana", "cherry", "date", "strawberry"] 

for fruit  in fruits : print(fruit) 

# Here as we can see that the "fruits" is the list and the word "fruit" is used as a variable name  
```

```python
# here we take another example,, where we have to print the characters of the whole name with spaces in between 

name = "Chaitanya Mangla"  
for ch in name :
    print(ch, end = ' ')  

# All the charcters are getting printed in a single  line with spaces in between 
```

For controlling the flow of loops, we use the 2 main logic keywords as :
1. Break 
2. Continue  

Break causes the termination of the loop and allows the execution of the code after the loop 
Continue allows next iteration of the loop to continue and run immediately just after the continue execution 

```python
# For example, the numbers from 1 to 12 have to be printed but as soon the number divisible by 4 is achieved, 
# we have to  break he code  

for num in range (1,13):
    if (num%4 == 0):
        break 
    print(num)

# Here as we can see in the code, the numbers are getting printed and the number 4 comes, the whole code stops
# Hence, the numbers get printed from 1 to 3 
```

```python
# Now, we have to print the number from 0 tp 10, but the condition is that the odd numbers should not get printed 

for i in range (0,11):
    if(i%2!=0):
        continue
    print(i)

# Here as we can see that the numbers which are odd don't get printed and only the even ones are printed using the continue statement
```

Nested Loops is the execution of the loops, which when executed the loop is printed inside the loop 


```python
# Nested Loop example takes place 
for i in range (10):
    for i in range (5):
        print("Hello World", end = " ")
```

```python
# Nested Loop Example takes place 
for i in range (10):
    for i in range (5):
        print("Hello World")
```

```python
# Example of nested for loop statements 
for i in range (4):
    for j in range(13):
        if (j%4 == 0):
            break
        print(j)

# Here nothing gets printed as because the number 0 is itself divisible by 4 and hence it breaks the statement 
```

```python
# Now e try the above code by starting the range from 1 
for i in range(4):
    for j in range(1,11):
        if (j%4 == 0):
            break
        print(j)

# here in the code as we can se that the loop gets executed and 1,2,3 gets printed 4 times 
```

```python
# Now we apply the above example to get even numbers printed 
for i in range (4):
    for j in range(1,13):
        if (j%2 != 0):
            continue
        print(j)
```

While Loop is used when we want to  keep running the loop until the condition is running. 

```python
num = 10 
while num > 10 :
    print("Hello World")

# Here since the number is equal to 10, so it will not print 10 as because condition is of num greater than 10 
```

```python
# While Loop execution as 
num = 10 
while num > 0 :
    print("Hello World")
    num = num - 1  

# here as we can see the the value is detting printed properly as because the while loop is xecuted properly 
```

```python
''' 
Write a python program to find the sum of all even numbers between 1 to 20 using for loop 
'''

sum = 0 
for num in range (1,21):
    if num%2 == 0 :
        sum += num 

print("Sum of even numbers is: ", sum)
```

```python
'''
Write a Python program to find whether a number is Prime or not 
''' 

num = int(input("Enter the number "))
isPrime = True

if (num <= 1):
    isPrime = False  

for i in range (2, int(num/2) + 1): 
    if num % i == 0 :
        isPrime = False 
        break 

print("Number passed is prime: ", isPrime) 

# The number input = 23 
```
